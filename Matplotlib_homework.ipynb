```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

city_data_to_load = "data/city_data.csv"                        # File to Load 
ride_data_to_load = "data/ride_data.csv"                        # File to Load

city_df = pd.read_csv(city_data_to_load)                        # Read the City and Ride Data
ride_df = pd.read_csv(ride_data_to_load)

merge_df = pd.merge(ride_df, city_df, on="city", how="outer")   # Combine the data into a single dataset                                    


merged_city_type = merge_df.set_index("type")                   # Set the 'type' to be our index 
new=merged_city_type.reset_index()                              #new merged database with clean index
new.head()                                                      #preview of the dataframe
```

```python
urban = new.loc[new["type"]== "Urban", :]          #splitting 'urban' from the frame
rural = new.loc[new["type"]== "Rural", :]          #splitting 'rural' from the frame
suburban = new.loc[new["type"]== "Suburban", :]    #splitting 'suburban' from the frame

rides_per_city_urban=urban.groupby("city").count().ride_id       #determining rides per city for urban type
rides_per_city_rural=rural.groupby("city").count().ride_id       #determining rides per city for rural type 
rides_per_city_suburban=suburban.groupby("city").count().ride_id #determining rides per city for suburban type

average_fare_city_urban=urban.groupby("city").mean().fare        #determining average fare for urban type
average_fare_city_rural=rural.groupby("city").mean().fare        #determining average fare for rural type
average_fare_city_suburban=suburban.groupby("city").mean().fare  #determining average fare for suburban type

drivers_by_city_urban=urban.groupby("city").sum().driver_count       #determining driver count for urban type
drivers_by_city_rural=rural.groupby("city").sum().driver_count       #determining driver count for rural type
drivers_by_city_suburban=suburban.groupby("city").sum().driver_count #determining driver count for suburban type

results_by_city_urban=pd.DataFrame({"Average Fare per City":average_fare_city_urban, #new dataframe for city type urban
                "Total Rides per City":rides_per_city_urban,
                 "Driver Count":drivers_by_city_urban})

results_by_city_rural=pd.DataFrame({"Average Fare per City":average_fare_city_rural, #new dataframe for city type rural
                "Total Rides per City":rides_per_city_rural,
                 "Driver Count":drivers_by_city_rural})

results_by_city_suburban=pd.DataFrame({"Average Fare per City":average_fare_city_suburban, #new dataframe for suburban
                "Total Rides per City":rides_per_city_suburban,
                 "Driver Count":drivers_by_city_suburban})
```

```python
ax=results_by_city_urban.plot(kind='scatter', x='Total Rides per City', y='Average Fare per City',
                           color='Red', label='Urban',edgecolor="black",alpha=0.5,grid=True,figsize=(8,6),
                           s=results_by_city_urban['Driver Count'],title="Pyber Ride Sharing Data 2016");
results_by_city_rural.plot(kind='scatter', x='Total Rides per City', y='Average Fare per City',
                           color='Yellow', label='Rural',edgecolor="black",ax=ax,alpha=0.5,grid=True,figsize=(8,6),
                           s=results_by_city_rural['Driver Count']);
results_by_city_suburban.plot(kind='scatter', x='Total Rides per City', y='Average Fare per City',
                           color='Blue', label='Suburban',edgecolor="black",ax=ax,alpha=0.5,grid=True,figsize=(8,6),
                           s=results_by_city_suburban['Driver Count'])
plt.show
```
